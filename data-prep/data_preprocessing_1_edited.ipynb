### Parsing joke text files

```python
# Imports
from bs4 import BeautifulSoup
import os
import re
```

```python
# Get a list with filenames
dir_path = r'../data/jester-data/jokes'
listd = os.listdir(dir_path)

# Remove .DS_Store file from list
listd = list(filter(lambda x : x != '.DS_Store', listd))

# Sort in asc order
listd.sort(key=lambda f: int(re.sub('\D', '', f)))
```

```python
# Compose a dict with joke texts
joke_text_raw = {}

for i in range(len(listd)):
    filename = listd[i]
    filename = '../data/jester-data/jokes/' + filename
    with open(filename) as fp:
        soup = BeautifulSoup(fp, "html.parser")

        # Get list of strings from html
        text_arr = [text for text in soup.stripped_strings]

        # Skip first elem, aka 'Joke [joke number here]'
        text = ''
        for j in range(len(text_arr)):
            if j == 0:
                continue
            else:
                text = text + text_arr[j]
        
        # Finally save to dict
        joke_text_raw[i] = text
```

### Composing final file

```python
# Imports
import json
```

```python
# Transform pd col with normalized ratings to list
ratings_list = train1['mean'].tolist()
```

```python
# Combine ratings and text
jester_json_list = []

for i in range(len(joke_text_raw)):
    joke_obj = {}

    joke_obj['body'] = joke_text_raw[i]
    joke_obj['rating'] = ratings_list[i]
    joke_obj['id'] = i

    jester_json_list.append(joke_obj)
```

```python
# Save to file jesterdata.json
with open(r'../data/jester-data/jesterdata.json', 'w') as output_file:
    json.dump(jester_json_list, output_file)
```
